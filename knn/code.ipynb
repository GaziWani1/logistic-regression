```python
import numpy as np

train_data = np.array([
    [2.0, 3.0, 0],   # ID1 -> A
    [3.0, 4.0, 1],   # ID2 -> A
    [5.0, 2.0, 1],   # ID3 -> B
    [6.0, 3.0, 1],   # ID4 -> B
    [7.0, 1.0, 1],   # ID5 -> B
    [4.0, 5.0, 0]    # ID6 -> A
])

X_train = train_data[:, :2]   
y_train = train_data[:, 2] 
J
print(X_train , y_train)
```

```python
x_test = np.array([4.5, 3.5])
```

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
def knn_classification(X_train, y_train, x_test, k=5):
    distances = [euclidean_distance(x_test, x_train) for x_train in X_train]

    # Returns sorted indices : We have to hold sorted index for labels
    k_indices = np.argsort(distances)[:k]


    # print(k_indices)

    k_nearest_labels = y_train[k_indices]

    # print(k_nearest_labels)

    count_0 = sum([1 for label in k_nearest_labels if label == 0])
    count_1 = sum([1 for label in k_nearest_labels if label == 1])

    # print("Count 0:", count_0)
    # print("Count 1:", count_1)

    # Prediction
    prediction = 1 if count_1 > count_0 else 0

    return distances , count_0 , count_1 , prediction
```

```python
distances , count_0 , count_1 , prediction = knn_classification(X_train, y_train, x_test, k=5)
print("Distances : " ,distances) 
print("Count 0 : " ,count_0)
print("Count 1 : ",count_1) 
print("Prediction : ",prediction)
```
