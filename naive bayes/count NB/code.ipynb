```python
import numpy as np

X = np.array([
    [1, 1],  # Yes Yes
    [1, 0],  # Yes No
    [0, 1],  # No Yes
    [1, 1],
    [0, 0],
    [0, 1],
    [1, 0],
    [0, 0]
])

y = np.array([1, 1, 0, 1, 0, 0, 1, 0])

x_test = np.array([1, 1])

total_samples = len(y)

disease_yes = np.sum(y == 1)
disease_no = np.sum(y == 0)

P_yes = disease_yes / total_samples
P_no = disease_no / total_samples

print("P(Disease=Yes):", P_yes)
print("P(Disease=No):", P_no)


X_yes = X[y == 1]
X_no = X[y == 0]


P_fever_yes_given_yes = np.sum(X_yes[:, 0] == 1) / len(X_yes)

P_cough_yes_given_yes = np.sum(X_yes[:, 1] == 1) / len(X_yes)


P_fever_yes_given_no = np.sum(X_no[:, 0] == 1) / len(X_no)

P_cough_yes_given_no = np.sum(X_no[:, 1] == 1) / len(X_no)

print("\nLikelihoods")

print("P(Fever=Yes | Disease=Yes):",
      P_fever_yes_given_yes)

print("P(Cough=Yes | Disease=Yes):",
      P_cough_yes_given_yes)

print("P(Fever=Yes | Disease=No):",
      P_fever_yes_given_no)

print("P(Cough=Yes | Disease=No):",
      P_cough_yes_given_no)


# Score for Disease = Yes
score_yes = (
    P_yes
    * P_fever_yes_given_yes
    * P_cough_yes_given_yes
)

# Score for Disease = No
score_no = (
    P_no
    * P_fever_yes_given_no
    * P_cough_yes_given_no
)

print("\nScores")

print("Disease Yes Score:", score_yes)
print("Disease No Score:", score_no)


if score_yes > score_no:
    prediction = 1
else:
    prediction = 0

print("\nPrediction:")

if prediction == 1:
    print("Patient HAS disease")
else:
    print("Patient DOES NOT have disease")
```
